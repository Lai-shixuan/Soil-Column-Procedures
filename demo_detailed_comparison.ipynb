```python
from API_functions import file_single as fs
from API_functions import file_compare as fc
from API_functions.Soils import threshold_position_independent as tpi
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
path = 'data/002_ou_DongYing_00001.bmp'
image = fs.read_image_opencv(path)
zoom_region = fc.zoom_region(x=210, y=250, width=200, height=100)

fig = plt.figure(figsize=(10, 6))
ax_zoom_in = fig.add_axes([0, 0.1, 0.3, 0.3])
ax_threshold_zoom_in = fig.add_axes([0, 0.6, 0.3, 0.3])
ax_origin = fig.add_axes([0.35, 0.1, 0.6, 0.8])


image_zoom_in = fc.zoom_in(image, zoom_region)
ax_zoom_in.imshow(image_zoom_in, cmap='gray')

threshold = 150
image_threshold = tpi.user_threshold(image, threshold)
image_threshold_zoom_in = fc.zoom_in(image_threshold, zoom_region)
ax_threshold_zoom_in.imshow(image_threshold_zoom_in, cmap='gray')

cv2.rectangle(image, zoom_region.cv_point1, zoom_region.cv_point2, (0, 0, 255), 3)
ax_origin.imshow(image, cmap='gray')
```
